# Data Interpolation

### 1. Linear Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python

import matplotlib.pyplot as plt
plt.scatter(x,y)
```

Here we can see the data is in linear fasion now if i want to add data points with in this points

##### Interpolate the data using linear interpolation

```python
x_new = np.linspace(1,5,10)  ##create new x values
y_interp = np.interp(x_new,x,y)  ##interpolatey values
print(y_interp)
```

```python

plt.scatter(x_new, y_interp)
```

Here we can see data points got added (10 data points)

### 2. Cubic interpolation with scipy

```python
import numpy as np

x = np.array([1,2,3,4,5])
y=np.array([1,4,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
##Create a cubic interpolation function
f = interp1d(x,y,kind = 'cubic')
```

```python
##Interpolate the data
x_new = np.linspace(1,5,10)
y_interp = f(x_new)
print(y_interp)
```

```python
# lets see data points before interpolation

plt.scatter(x,y)
```

```python
## Data points after interpolation

plt.scatter(x_new, y_interp)
```

Here we can see new data ponits got added 

### 3. Polynomial Interpolation

```python
#create some sample data

x = np.array([1,2,3,4,5])
y= np.array([2,4,6,8,10])
```

```python
#Interpolate the data using polynomial interpolation

p = np.polyfit(x, y, 2) #fit a 2nd degree polynomial to the data

#Degree value tells how much to bend the data points
```

```python
x_new = np.linspace(1, 5, 10) # create new x values
y_interp = np.polyval(p, x_new) # interpolate y values
```

```python
# graph before interpolation
plt.scatter(x, y)
```

```python
# graph after interpolation
plt.scatter(x_new, y_interp)
```

Here we can see the new data points got added
